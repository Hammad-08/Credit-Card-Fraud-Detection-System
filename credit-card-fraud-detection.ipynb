```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Random Forest Classifier": RandomForestClassifier(),
    "Support Vector Classifier": SVC(),
    "K-Nearest Neighbors Classifier": KNeighborsClassifier(),
    "Gaussian Naive Bayes": GaussianNB(),
    "AdaBoost Classifier": AdaBoostClassifier(),
    "Gradient Boosting Classifier": GradientBoostingClassifier(),
    "Bagging Classifier": BaggingClassifier(),
    "Extra Trees Classifier": ExtraTreesClassifier(),
    "Stochastic Gradient Descent Classifier": SGDClassifier(),
    "Voting Classifier": VotingClassifier(estimators=[
        ('lr', LogisticRegression()),
        ('dt', DecisionTreeClassifier()),
        ('rf', RandomForestClassifier()),
        ('svc', SVC()),
        ('knn', KNeighborsClassifier())
    ], voting='hard')
}

for name, clf in classifier.items():
    print(f"\n=========={name}===========")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    # Evaluation Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    print(f"\n Accuracy: {accuracy}")
    print(f" Precision: {precision}")
    print(f" Recall: {recall}")
    print(f" F1 Score: {f1}")
    
    # Confusion Matrix
    print("\n Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    # Classification Report
    print("\n Classification Report:")
    print(classification_report(y_test, y_pred))
```

```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
import joblib

# Load your trained model
model = joblib.load('credit_card_model.pkl')

def predict():
    # Get values from entry fields
    values = [float(entry.get()) for entry in entries]
    
    # Perform prediction
    prediction = model.predict([values])[0]
    if prediction == 0:
        result = "normal transaction"
    else:
        result = "fraud transaction"
    messagebox.showinfo("Prediction Result", f"The prediction is: {result}")

# Create tkinter window
window = tk.Tk()
window.title("Credit Card Fraud Detection System")
window.geometry("500x400")

# Set dark theme colors
bg_color = "#2c3e50"  # Background color
fg_color = "#ecf0f1"  # Foreground color
btn_bg_color = "#3498db"  # Button background color
btn_fg_color = fg_color  # Button foreground color

# Set background color for window
window.configure(bg=bg_color)

# Custom font
custom_font = ("Helvetica", 12, "bold")

# Title label
title_label = tk.Label(window, text="Credit Card Fraud Detection System", font=("Helvetica", 16, "bold"), bg=bg_color, fg=fg_color)
title_label.grid(row=0, column=0, columnspan=2, pady=10)

# Default values for each feature
default_values = [-1.3598071336738, 0.0727811733098497, 1.53634673796914, 1.37815522427443, -1.338320769942518, -1.462387777762292, 0.239598554061257, 1.0986979012610507, 0.363786969611213, -0.0907941719789316]

# Create entry fields for each feature with default values
feature_names = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 'Feature6', 'Feature7', 'Feature8', 'Feature9', 'Feature10']
entries = []
for i, (feature_name, default_value) in enumerate(zip(feature_names, default_values), start=1):
    label = tk.Label(window, text=feature_name, font=custom_font, bg=bg_color, fg=fg_color)
    label.grid(row=i, column=0, pady=5, padx=5, sticky="w")
    default_var = tk.StringVar(value=default_value)
    entry = tk.Entry(window, textvariable=default_var, font=custom_font, bg=fg_color, fg=bg_color)
    entry.grid(row=i, column=1, pady=5, padx=5)
    entries.append(entry)

# Create predict button
predict_button = tk.Button(window, text="Predict", command=predict, font=custom_font, bg=btn_bg_color, fg=btn_fg_color)
predict_button.grid(row=len(feature_names)+1, column=0, columnspan=2, pady=10)

# Add some color and interactivity to the window
window.option_add("*TButton.Background", btn_bg_color)
window.option_add("*TButton.Foreground", btn_fg_color)
window.option_add("*Font", custom_font)

# Run tkinter event loop
window.mainloop()
```
